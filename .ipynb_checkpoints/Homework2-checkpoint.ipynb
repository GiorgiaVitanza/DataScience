```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

Homework 1:

```python
df = pd.read_csv('df3.csv')
```

```python
col = list(df.columns)
```

```python
plt.figure(figsize=(9,5))
corr = df[col].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=1)
plt.title("Correlation Heatmap")
```

The columns that contributed to create the column "f" are the columns "b" and "y" with a negative correlation and "d" with a positive correlation.
Let's have a look at the scatter plots of these columns:

```python
plt.scatter(df['f'],df['b'])
plt.xlabel('f')
plt.ylabel('b')

plt.show()

plt.scatter(df['f'],df['y'])
plt.xlabel('f')
plt.ylabel('y')
plt.show()


plt.scatter(df['f'],df['d'])
plt.xlabel('f')
plt.ylabel('d')
plt.show()
```

We can conclude that also the scatter plots confirm the correlation between the already said columns "d", "y" and "b". In particular the "f" variable is splitted into two groups in all the cases above.

Homework 2: Plot y = 1/x between -10 and 10

```python
x_left = np.linspace(-10,-0.01,50)
x_right = np.linspace(0.01,10,50)
y_left = 1/x_left
y_right = 1/x_right

plt.plot(x_left,y_left,color = 'blue')
plt.plot(x_right,y_right, color = 'blue')
plt.show()
```

Homework 3: Recreate the heatmaps:

```python
x_vals = np.linspace(-10, 10, 50)
y_vals = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x_vals, y_vals)
Z = X*Y  # iperbole

df = pd.DataFrame({'x':X.ravel(), 'y':Y.ravel(), 'z':Z.ravel()})
heatmap1 = df.pivot(index = 'y', columns = 'x', values = 'z') 
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap1, cmap='coolwarm', annot=False)
plt.title('Heatmap of z as function of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_vals = np.linspace(-10, 10, 50)
y_vals = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x_vals, y_vals)
Z = X**2+Y**2  # circonferenza

df = pd.DataFrame({'x':X.ravel(), 'y': Y.ravel(), 'z':Z.ravel()})
heatmap2 = df.pivot(index = 'y', columns = 'x', values = 'z')
# Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(heatmap2, cmap='coolwarm', annot=False)
plt.title('Heatmap of z as function of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_vals = np.linspace(-10, 10, 50)
y_vals = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x_vals, y_vals)
Z = X**2+Y  # parabola
# Convert to DataFrame
df_simulation = pd.DataFrame({'x': X.ravel(), 'y': Y.ravel(), 'z': Z.ravel()})

# Pivot the data for heatmap
heatmap_data = df_simulation.pivot(index='y', columns='x', values='z')
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=False)
plt.title('Heatmap of z as function of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
