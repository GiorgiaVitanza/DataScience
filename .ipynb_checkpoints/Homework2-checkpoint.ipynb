```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(np.linspace(0.001, 10,15),np.linspace(10, 50,15))
y = np.log(x)

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define grid size
size = 100
x = np.linspace(-1, 1, size)
y = np.linspace(-1, 1, size)
X, Y = np.meshgrid(x, y)

# Define heatmaps
heatmap1 = X  # Horizontal gradient
heatmap2 = X + Y  # Diagonal gradient


# Plot first heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(heatmap1, cmap='coolwarm', annot=False)
plt.title('Vertical Heatmap')
plt.xlabel('x')
plt.ylabel('y')
# Plot second heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(heatmap2, cmap='coolwarm', annot=False)
plt.title('Diagonal Heatmap')
plt.xlabel('x')
plt.ylabel('y')
plt.show()



```
