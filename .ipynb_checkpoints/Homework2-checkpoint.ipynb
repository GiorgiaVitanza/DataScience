```python
import numpy as np
import matplotlib.pyplot as plt

# Define grid size
size = 100
x = np.linspace(-1, 1, size)
y = np.linspace(-1, 1, size)
X, Y = np.meshgrid(x, y)

# Define heatmaps
heatmap1 = X  # Horizontal gradient
heatmap2 = X - Y  # Diagonal gradient

# Create figure
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot first heatmap
c1 = axes[0].imshow(heatmap1, cmap='coolwarm', origin='lower', aspect='auto')
fig.colorbar(c1, ax=axes[0])
axes[0].set_title("Horizontal Gradient")

# Plot second heatmap
c2 = axes[1].imshow(heatmap2, cmap='coolwarm', origin='lower', aspect='auto')
fig.colorbar(c2, ax=axes[1])
axes[1].set_title("Diagonal Gradient")

# Show plots
plt.tight_layout()
plt.show()

```
